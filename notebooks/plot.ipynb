```python
import matplotlib.pyplot as plt
import numpy as np

# Data organized by judge pairs
judge_pairs = ['OpenAI Judge', 'Gemini Judge']

# System scores for each judge
systems_data = {
    'Perplexity Assistant': [11.43, 31.42],
    'Gemini Assistant': [18.10, 27.62], 
    'VertexAI Assistant': [78.10, 85.70],
    'OpenAI Assistant': [80.00, 88.57],
    'OpenAIPDF Assistant': [84.76, 90.47],
    'PaperQA2': [65.05, 76.70],
    'Modified PaperQA2': [65.71, 81.90],
    'HybridGemGem Assistant': [80.00, 85.72],
    'HybridOAIGem Assistant': [83.65, 90.48]
}
# Define colors for each system - matching the JudgeBench style
system_colors = {
    'Perplexity Assistant': '#1f77b4',      # Blue
    'Gemini Assistant': '#ff7f0e',          # Orange
    'VertexAI Assistant': '#2ca02c',        # Green (OpenAI family)
    'OpenAI Assistant': '#d62728',          # Red (OpenAI family) 
    'OpenAIPDF Assistant': '#9467bd',       # Purple (OpenAI family)
    'PaperQA2': '#8c564b',                  # Brown
    'Modified PaperQA2': '#e377c2',         # Pink
    'HybridGemGem Assistant': '#7f7f7f',    # Gray (Hybrid)
    'HybridOAIGem Assistant': '#bcbd22'     # Olive (Hybrid)
}



# Set up the plot - similar proportions to JudgeBench
fig, ax = plt.subplots(figsize=(12, 8))

# X positions for judge pairs
x = np.arange(len(judge_pairs))
width = 0.8  # Total width for all bars in each group
n_systems = len(systems_data)
bar_width = width / n_systems

# Create grouped bars
for i, (system_name, scores) in enumerate(systems_data.items()):
    x_pos = x + (i - (n_systems-1)/2) * bar_width
    bars = ax.bar(x_pos, scores, bar_width, 
                  label=system_name, 
                  color=system_colors[system_name],
                  alpha=0.7,
                  edgecolor='white',
                  linewidth=0.5)
    
    # Add value labels on bars
    for bar, score in zip(bars, scores):
        height = bar.get_height()
        ax.annotate(f'{score:.1f}',
                   xy=(bar.get_x() + bar.get_width() / 2, height),
                   xytext=(0, 3),
                   textcoords="offset points",
                   ha='center', va='bottom',
                   fontsize=9, fontweight='bold')



# Customize the plot to match JudgeBench style
ax.set_ylabel('Accuracy (%)', fontsize=14, fontweight='bold')
ax.set_ylim(0, 100)
ax.set_yticks(range(0, 101, 10))

# Set x-axis
ax.set_xticks(x)
ax.set_xticklabels(judge_pairs, fontsize=14, fontweight='bold')

# Add legend below the plot like JudgeBench
legend = ax.legend(bbox_to_anchor=(0.5, -0.15), loc='upper center', 
                  ncol=3, fontsize=11, frameon=True)
legend.get_frame().set_facecolor('white')
legend.get_frame().set_alpha(0.9)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_linewidth(1)

# Add subtle grid
ax.grid(True, alpha=0.2, axis='y')
ax.set_axisbelow(True)


# Adjust layout to accommodate legend
plt.tight_layout()
plt.subplots_adjust(bottom=0.25)

plt.show()

# Print ranking comparison
print("\n" + "="*80)
print("SYSTEM RANKINGS COMPARISON")
print("="*80)

# Sort systems by each judge
openai_rankings = sorted(systems_data.items(), key=lambda x: x[1][0], reverse=True)
gemini_rankings = sorted(systems_data.items(), key=lambda x: x[1][1], reverse=True)

print(f"\n{'Rank':<4} {'OpenAI Judge':<25} {'Score':<6} {'Gemini Judge':<25} {'Score'}")
print("-" * 70)

for i in range(len(systems_data)):
    openai_system, openai_scores = openai_rankings[i]
    gemini_system, gemini_scores = gemini_rankings[i]
    print(f"{i+1:<4} {openai_system:<25} {openai_scores[0]:<6.1f} {gemini_system:<25} {gemini_scores[1]:.1f}")

print("\n" + "="*80)
```
